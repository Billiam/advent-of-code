```ruby
data = File.read('12.txt').split("\n").map {|i| [i[0], i[1..].to_i] }
```

```ruby
def add_vector(a, d)
  rad = a * Math::PI / 180
  @ship.x += (Math.cos(rad) * d).to_i
  @ship.y += (Math.sin(rad) * d).to_i
end
```

```ruby
@ship = OpenStruct.new({
  dir: 0,
  x: 0,
  y: 0
})

data.each do |cmd, dist|
  case cmd
    when "F"
     add_vector(@ship.dir, dist)
    when "N"
      add_vector(-90, dist)
    when "S"
      add_vector(90, dist)
    when "E"
      add_vector(0, dist)
    when "W"
      add_vector(180, dist)
    when "L"
      @ship.dir = (@ship.dir - dist) % 360
    when "R"
      @ship.dir = (@ship.dir + dist) % 360
  end
end
@ship.x.abs + @ship.y.abs
```

```ruby
require 'matrix'

@ship = OpenStruct.new({
  pos: Vector[0,0]
})
@waypoint =  OpenStruct.new({
  pos: Vector[10,1]
})

def move_to_waypoint(dist)
  move = (@waypoint.pos - @ship.pos) * dist
  @ship.pos += move
  @waypoint.pos += move
end

def move_waypoint(x,y)
  @waypoint.pos += Vector[x,y]
end

def rotate_waypoint(angle)
  offset = @waypoint.pos - @ship.pos
  # convert into left hand rotations (vector.cross_product)
  r = ((-1 * angle + 360) % 360) / 90
  r.abs.times { offset = offset.cross }
  @waypoint.pos = @ship.pos + offset 
end

data.each do |cmd, dist|
  case cmd
    when "F"
     move_to_waypoint(dist)
    when "N"
      move_waypoint(0, dist)
    when "S"
      move_waypoint(0, -dist)
    when "E"
      move_waypoint(dist, 0)
    when "W"
      move_waypoint(-dist, 0)
    when "L"
      rotate_waypoint(-dist)
    when "R"
      rotate_waypoint(dist)
  end
  
#   puts ({ cmd: [cmd, dist],
#     ship: @ship.pos,
#     way: @waypoint.pos,
#     diff: @waypoint.pos - @ship.pos,
#     }.inspect)
end
@ship.pos.to_a.map{|i| i.abs }.inject(&:+)
```
