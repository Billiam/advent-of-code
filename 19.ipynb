# Day 19

```ruby
data = File.read('19.txt').split("\n\n").map {|i| i.split("\n") }
```

```ruby
@rules = data.first.map {|i| i.match(/(\d+): (.*)/).captures}.map{|k,v| [k, v.split("|").map{|x| x.split(" ").reject(&:empty?).compact.map{|y| y.gsub('"', '')} }]}.sort_by(&:first).to_h
```

```ruby
lines = data[1]
```

```ruby
@cache = {}
def rule_map(key)
  @cache[key] ||= begin
    either_or = @rules[key]
    
    return key unless either_or
    pair = either_or.map do |group|
      group.map do |sub_rule|
        rule_map(sub_rule)
      end.join("")
    end.join("|")
    if pair.include?("|")
      "(" + pair + ")"
    else
      pair
    end
  end
end
rule_regex = Regexp.new("^#{rule_map("0")}$")
```

```ruby
lines.select {|n| rule_regex.match(n) }.count
```

## Part 2

```ruby
data2 = File.read('19b.txt').split("\n\n").map {|i| i.split("\n") }
```

```ruby
@rules2 = data2.first.map {|i| i.match(/(\d+): (.*)/).captures}.map{|k,v| [k, v.split("|").map{|x| x.split(" ").reject(&:empty?).compact.map{|y| y.gsub('"', '')} }]}.sort_by(&:first).to_h
```

```ruby
@cache2 = {}
def rule_map2(key, l=0)

  @cache2[key] ||= begin
    either_or = @rules2[key]
    
    l += 1
    return "" if l == 96
    return key unless either_or
    
    pair = either_or.map do |group|
      group.map do |sub_rule|
        rule_map2(sub_rule, l)
      end.join("")
    end.join("|")
    if pair.include?("|")
      "(?:" + pair + ")"
    else
      pair
    end
  end
end
rule_regex2 = Regexp.new("^#{rule_map2("0")}$")
data2[1].select {|l| rule_regex2.match(l) }.count
```
