# Day 11

```ruby
@seats = File.read("11.txt").split("\n").map {|i| i.split("").map {|i| i=="L"}}
```

```ruby
def run(state)
  new_state = Array.new(@seats.size) { Array.new(@seats.first.size) { false }}
  
  directions = [-1,0,1].repeated_permutation(2).reject {|d| d == [0,0] }

  @seats.each_with_index do |row_data, row|
    row_data.each_with_index do |seat, column|
      next unless seat
      adjacent = directions.select do |r,c|        
        (row+r).between?(0, @seats.first.size - 1) && (column+c).between?(0, @seats.first.size - 1)
      end.map { |r,c| state.dig(row+r, column+c) }.count(&:itself)        

      if state.dig(row,column) && adjacent >= 4
        new_state[row][column] = false
      elsif !state.dig(row,column) && adjacent == 0
        new_state[row][column] = true
      else
        new_state[row][column] = state.dig(row,column)
      end
    end
  end
  new_state
end


previous_state=[]
while true do
  current = run(previous_state)
  break current.count(&:itself) if current.flatten == previous_state.flatten
  puts current.map {|i| i.map {|n| n ? "#" : "."}.join("")}.join("\n")
  IRuby::Display.clear_output(true)
  previous_state = current
end

previous_state.flatten.count(&:itself)
```

## Part 2

```ruby
def run_distant(state)
  directions = [-1,0,1].repeated_permutation(2).reject {|d| d == [0,0] }
  new_state = Array.new(@seats.size) { Array.new(@seats.first.size) { false }}
  
  @seats.each_with_index do |row_data, row|
    row_data.each_with_index do |type, column|
      next unless type
      
      adjacent = directions.map do |r,c|
        look_dist = 1
        while true do
          r1 = r * look_dist + row
          c1 = c * look_dist + column
          
          break false if !r1.between?(0, @seats.first.size - 1) ||
            !c1.between?(0, @seats.first.size - 1)
          break true if state.dig(r1, c1)
          break false if @seats.dig(r1, c1)

          look_dist+=1
        end
      end.count(&:itself)

      if state.dig(row,column) && adjacent >= 5
        new_state[row][column] = false
      elsif !state.dig(row,column) && adjacent == 0
        new_state[row][column] = true
      else
        new_state[row][column] = state.dig(row,column)
      end
      
    end
  end
  new_state
end

previous_state=[]
while true do
  current = run_distant(previous_state)
  break current.count(&:itself) if current.flatten == previous_state.flatten
  puts current.map {|i| i.map {|n| n ? "#" : "."}.join("")}.join("\n")
  IRuby::Display.clear_output(true)
  previous_state = current
end

previous_state.flatten.count(&:itself)
```
