# Day 18

```ruby
data = File.read('18.txt').split("\n").map {|n| n.gsub(' ','') }
```

```ruby
data.map do |v|
  s = StringScanner.new(v)
  stack = [0]
  mod = nil
  
  until s.eos?
    case
      when open = s.scan('(')
        stack << (mod || '+')
        stack << 0
        mod = nil
      when s.scan(')')
        last_mod, paren_result = stack.pop(2)
        stack[-1] = stack.last.send(last_mod.to_sym, paren_result)
      when m = s.scan(/[+-\/*]/)
        mod = m
      when n = s.scan(/\d+/)
        if mod
          stack[-1] = stack.last.send(mod.to_sym, n.to_i)
        else
         stack[-1] += n.to_i
       end
       mod = nil
    else
    end
  end
  stack.first
end.inject(:+)
```

## Part 2

It's not dumb if it works

```ruby
data.map do |v|
  v=v.gsub(" ","")
 
  loop do
    ov = v
    # evaluate strings of a+b+c
    v = v.gsub(/\d+(\+\d+)+/) do |r|
      eval(r)
    end
    
    # evaluate top level paren groups
    v = v.gsub(/\([^())]+\)/) do |r|
      eval(r)
    end
 
    break if ov == v
  end

  eval(v)
end.inject(:+)
```
