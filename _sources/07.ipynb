# Day 7

```ruby
data = File.read('7.txt').split("\n");nil
```

```ruby
containers = data.map do |d|
  key, val = d.split(' contain ')
  key=key[0..-6]
  vals = val.split(", ").map {|bag| m=/(\d+) (.*) bag/.match(bag); m && m[2] }
  [key, vals]
end.to_h
```

```ruby
parent_containers = {}
containers.each do |k,v|
  v.each do |bag|
    parent_containers[bag] ||= []
    parent_containers[bag] << k
  end
end;nil
```

```ruby
def find_parents(container, type)
  Array(container[type]).map do |parent|
    [parent] + find_parents(container, parent)
  end
end
```

```ruby
found = find_parents(parent_containers, 'shiny gold').flatten.uniq
```

```ruby
found.count
```

```ruby
sized_containers = data.map do |d|
  key, val = d.split(' contain ')
  key=key[0..-6]
  vals = val.split(", ").map {|bag| m=/(\d+) (.*) bag/.match(bag); m && [m[2],m[1].to_i] }
  [key, vals] if vals.any?
end.compact.to_h;nil
```

```ruby
def bag_count(container, type, count=1)
  Array(container[type]).map do |child_bag, number|
    count * (number + bag_count(container, child_bag, number))
  end.sum
end
bag_count(sized_containers, 'shiny gold')
```
